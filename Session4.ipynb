```python
import requests as req

url_target = 'https://opentutorials.org/course/1'
res_get = req.get(url_target)

from bs4 import BeautifulSoup

soup = BeautifulSoup(res_get.text, "html.parser")

import pandas as pd
li_name = soup.select('div.name.time > strong')
li_name_processed = [ e.text.strip() for e in li_name ]
li_time = soup.select('div.name.time > a > time')
li_time_processed = [ e.text.strip() for e in li_time ]
li_comment = soup.select(' div.comment_content')
li_comment_processed = [ e.text.strip() for e in li_comment ]

container = {'name' : li_name_processed, 'time': li_time_processed, 'comment' : li_comment_processed}
df_table = pd.DataFrame(container)
df_table
```
